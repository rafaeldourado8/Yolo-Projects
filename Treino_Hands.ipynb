```python
from ultralytics import YOLO
import os

# --- PASSO 1: DESCOMPACTAR ---
# O nome original com caracteres especiais
ZIP_FILE_NAME = 'Hand.v2i.yolov8 (1).zip'

# A MÁGICA ESTÁ AQUI: Colocamos aspas "{ZIP_FILE_NAME}"
# Isso força o terminal a ler o nome do arquivo inteiro
!unzip -o "{ZIP_FILE_NAME}" -d dataset_gestos/

# --- PASSO 2: DEFINIR O CAMINHO ---
data_yaml_path = 'dataset_gestos/data.yaml'

# --- PASSO 3: VERIFICAR E TREINAR ---
if not os.path.exists(data_yaml_path):
    print(f"ERRO: Arquivo 'data.yaml' não foi encontrado em '{data_yaml_path}'")
    print(f"Verifique se o arquivo {ZIP_FILE_NAME} foi descompactado corretamente.")
else:
    print(f"Arquivo '{data_yaml_path}' encontrado. Iniciando o treinamento...")

    model = YOLO('yolov8n.pt')

    try:
        results = model.train(
            data=data_yaml_path,
            epochs=50,
            imgsz=640,
            name='yolo_gestos_v1'
        )
        print("\n--- Treinamento Concluído! ---")
    except Exception as e:
        print(f"\nOcorreu um erro durante o treinamento: {e}")
```
